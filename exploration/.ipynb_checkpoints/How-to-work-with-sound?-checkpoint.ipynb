### 0.2 Write audio

Suppose we successfuly "predict" a numpy array of an ouptut song. Can we convert it into an `mp3` file? 
(Hint: yes.)

```python
# create a sign wave signal
sig_sin = (np.sin(np.arange(441000)/40)*1e4).astype(np.int16)
```

```python
audiosegment.from_numpy_array(sig_sin, framerate=88200.0)
```

```python
# to write this segment, we'll have to convert it to an AudioSegment object
def toAudioSegment(arr,framerate):
    seg = AudioSegment(
    arr.tobytes(), 
    frame_rate=framerate,
    sample_width=arr.dtype.itemsize, 
    channels=1
    )
    return seg
```

```python
segAudio = toAudioSegment(sig_sin, 88200)
segAudio
```

```python
# write this sin wave to a file. 
segAudio.export("output/sample.mp3", "mp3")
```

## 1. Normalizing audio

Say we have 5 second clips. We want a standard:

- minimum/maximum amplitude
- Round to a `np.int16` datatype

Here's a function to do just that, as well as return the framerate

```python
def normalize_amplitude(arr, seconds, minamp = -1e4, maxamp = 1e4):
    # calculate frames per seconds
    framerate = len(arr)/seconds
    # shift min of 0
    arr = arr - arr.min()
    # scale up so that range is 20000
    arr = arr*(2e4/arr.max())
    # shift everything left 10k
    arr = arr - 1e4
    # convert to integer 
    arr = arr.astype(np.int16)
    
    return arr, framerate
```

```python
sig_norm, sig_framerate = normalize_amplitude(sig_sin, seconds=5)
print(sig_norm)
print(sig_framerate)
```

## 2. Creating spectograms



```python
def normalize_spectogram(Sxx):
    Sxx = np.log10(Sxx + 1e-9)
    smin = Sxx.min()
    srange = Sxx.max() - smin
    Sxx = (Sxx - smin)/srange
    # reshape to a (662, 666) image
    # Sxx = resize(Sxx,(662, 666), mode = "constant")
    return Sxx
```

```python
def get_amplitudes(seg):
    freqs, times, amplitudes = seg.spectrogram(window_length_s=0.03, overlap=.9)
    return freqs, times, amplitudes
```

```python
seg = audiosegment.from_numpy_array(sig_norm, framerate=sig_framerate)

freqs, times, amplitudes = get_amplitudes(seg)
print(amplitudes.shape)
amplitudes = normalize_spectogram(amplitudes)
# Plot
fig = plt.figure(figsize=(12,10))
plt.pcolormesh(times, freqs, amplitudes)
plt.xlabel("Time in Seconds")
plt.ylabel("Frequency in Hz")
plt.show()

```

```python
# first 1 seconds o "somewhere over the rainbow" (excluding the first seconds)
seg1 = songs2[1][1*1000:2*1000]

freqs, times, amplitudes1 = get_amplitudes(seg1)
amplitudes1 = normalize_spectogram(amplitudes1)
# Plot
fig = plt.figure(figsize=(12,10))
plt.pcolormesh(times, freqs, amplitudes1)
plt.xlabel("Time in Seconds")
plt.ylabel("Frequency in Hz")
plt.show()

print(amplitudes1.shape)
print(times.shape)
print(freqs.shape)
print(min(freqs), max(freqs))
```
